```python
penguins = sns.load_dataset("penguins")

# Create a scatter plot with different colors for each species
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
)

# Add labels and title
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.title("Scatter Plot of All Penguin Species")

# Show the plot
plt.show()
```

# For species i want Adelie and group by island and sex

```python
adelie_data = penguins[penguins['species'] == 'Adelie']

# Group the data by island
grouped_data = adelie_data.groupby('island')

# Print the grouped data as a table
print(grouped_data.describe())

# Create a scatter plot
plt.figure(figsize=(8, 6))

for name, group in grouped_data:
    plt.scatter(group['bill_length_mm'], group['bill_depth_mm'], label=name)

plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend()
plt.show()
```
